```python
import pandas as pd 
df=pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/Exercise/titanic.csv')
```

```python
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
```

```python
inputs =df.drop(['Survived'],axis ='columns')
traget =df['Survived']
```

```python
y=traget
y
```

```python
inputs.Sex = inputs.Sex.map({'male': 1, 'female': 2})
```

```python
inputs.Age[:10]
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
```

```python
inputs.head()
```

```python
inputs.head()
```

```python
x=inputs
x
```

```python
from sklearn.model_selection import train_test_split 
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.1)
```

```python
len(x_test)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
#finally we got our solution 
model.predict([[2,1,28,18]])
```

```python
#second method for predicting the titanic passenger list 
from sklearn.linear_model import LinearRegression 
ks=LinearRegression()
```

```python
ks.fit(x_train,y_train)
```
